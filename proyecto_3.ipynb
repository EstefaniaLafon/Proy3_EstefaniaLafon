# INTRODUCCIÓN
La tasa de natalidad es un indicador demográfico que expresa la frecuencia con que ocurren los nacimientos en la población y se expresa como el número de nacimientos por cada mil habitantes. 
Para calcular la misma se utiliza la siguiente fórmula:
* Número de Nacimientos es la cantidad de bebés nacidos en un período dado.
* Población Total es el número total de habitantes de la misma área durante el mismo período.

**Tasa de Natalidad = (Número de Nacimientos / Población Total) * 1000**

El conocimiento de este indicador demográfico es importante por varias razones: 
- **Indica el crecimiento de la población:** Cuando la tasa es alta sugiere un crecimiento poblacional, mientras que una tasa baja puede indicar un estancamiento o disminución de la población.
- **Impacto en la estructura de edad:** Una alta tasa de natalidad puede llevar a una población más joven
- **Planificación y políticas públicas:** Los gobiernos y las instituciones utilizan la tasa de natalidad para planificar recursos y servicios que se necesitarán en el futuro, como escuelas, hospitales y programas de bienestar.
- **Influencia en la economía:** La tasa de natalidad puede tener un impacto en la economía, ya que una población más grande y joven puede estimular el consumo y la producción, pero también puede aumentar la necesidad de empleos y servicios públicos.

## TASA NATALIDAD POR AÑO DE ARGENTINA Y URUGUAY
* Region: Uruguay y Argentina
* Categoria: Natalidad

**¿Cómo ha evolucionado la tasa de natalidad en Uruguay en comparación con Argentina entre los años 2000 y 2017?**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

#Descargar los dataset
tasa_uy_total = pd.read_csv("../Proy3_EstefaniaLafon/data/tasanatalidad_totalpaisury.csv",delimiter=",",header=0,encoding="latin1")
tasa_ar = pd.read_csv("../Proy3_EstefaniaLafon/data/AR_tasa_natalidad.csv",delimiter=",",header=0,encoding="latin1")

#Convertir los años
tasa_ar["indice_tiempo"]= pd.to_datetime(tasa_ar["indice_tiempo"])
tasa_ar["año"] = tasa_ar["indice_tiempo"].dt.year.astype(int)  # Asegurarse de que sea un entero

#Filtrar los datos de cada dataset para que sean comparables
tasa_uy_filtrado=tasa_uy_total[(tasa_uy_total["año"]>=2000)&(tasa_uy_total["año"]<=2017)]
tasa_ar_filtrado=tasa_ar[(tasa_ar["año"]>=2000)&(tasa_ar["año"]<=2017)]

#Armado del gráfico
plt.title("Tasa Natalidad Uruguay - Argentina")
plt.xlabel("Año")
plt.ylabel("Tasa Natalidad")


plt.plot(tasa_uy_filtrado["año"], tasa_uy_filtrado["valor"], label="Uruguay", color="blue")
plt.plot(tasa_ar_filtrado["año"], tasa_ar_filtrado["natalidad_argentina"], label="Argentina", color="red")


plt.legend(loc='upper right')

plt.gca().xaxis.set_major_formatter(mtick.FormatStrFormatter('%d'))

plt.show()
```

# **JUSTIFICACION**

A partir del código proporcionado se visualiza un gráfico el cual compara la tasa de natalidad de Uruguay en comparacion con Argentina desde los años 2000 a 2017.
El gráfico generado muestra dos lineas distintas, la linea azul representa Uruguay y la linea roja representa a Argentina. Como se puede visualizar, si bien la tasa de natalidad de Argentina es mayor a la de Uruguay, podemos decir que ambas vienen disminuyendo en el paso de los años. Esto implica para ambos paises una disminución de crecimiento de la población para ambos paises en un futuro, aunque la disminucion de la poblacion Argentina será menor que la de Uruguay.

El uso de la biblioteca Matplotlib para crear el gráfico garantiza que la información sea presentada de manera clara y atractiva, siguiendo el principio de belleza de Cairo. La incorporación de una leyenda en la esquina superior derecha del gráfico agrega valor a la visualización al proporcionar una guía visual para identificar qué color corresponde a cada país.

En resumen, el código cumple con los principios de veracidad, funcionalidad, belleza y perspicacia de Cairo al brindar una representación precisa y efectiva de los datos para responder a la pregunta de investigación de manera clara y accesible.
